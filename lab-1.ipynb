### 1 - Add a new column : Salary_class 

```python
dataset.insert(4, "Salary_class", "nan")
dataset
```

### 2-  A for loop is implemented and the observations are separated into three categories: 
#### Salary 
- greater than 70000 -  class0
- between 61000-70000 - class1
- between  48000-60000 -class2
- The classes have been stored in a new column ‘Salary Class’

```python
for i in range(len(dataset)):
    if dataset["Salary"][i] > 70000:
        dataset.Salary_class[i]="class0"
    elif dataset["Salary"][i] >= 61000 and dataset["Salary"][i] <= 70000:
        dataset.Salary_class[i]="class1"
    elif dataset["Salary"][i] >= 48000 and dataset["Salary"][i] <= 60000:
        dataset.Salary_class[i]="class2"

dataset
```

### 3 -	Count the number of each class (class 0, class1,class2) in your dataset.

```python
i = 1
cnt0=0 
cnt1=0 
cnt2=0
while i < len(dataset):
    if dataset.Salary_class[i] == "class0":
        cnt0 += 1
    elif dataset.Salary_class[i] == "class1":
        cnt1 += 1
    elif dataset.Salary_class[i] == "class2":
        cnt2 += 1
    i += 1
print(cnt0, cnt1, cnt2)
```
