```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 1. Plot a histogram with blue color bars of size 4, and edges should be distinguished from each other, for the dataset social_nework for the feature estimated salary.

```python
dataset = pd.read_csv("social.csv")

plt.hist(dataset['EstimatedSalary'],color = 'blue',edgecolor = 'white',bins = 4)
plt.title('Histogram of Social Network')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.show()
```

### 2. On the dataset ‘data’, draw barplot to show the count of categorical feature ‘Country’

```python
dataset = pd.read_csv("Data.csv")
counts= [3, 3, 3]

countryName = ('France','Spain','Germany')
index = np.arange(len(countryName))

plt.bar(index, counts, color=['red','blue', 'cyan' ])
plt.title('Bar plot of Country')
plt.xlabel('Country')
plt.ylabel('Frequency')
plt.show()
```

### 4. Scatter polt age vs estimated salary on gridview
- A)	show regression fit line

```python
dataset = pd.read_csv("social.csv")
sns.set(style="darkgrid")
sns.regplot(x=dataset['Age'],y=dataset['EstimatedSalary'])
```

- B) Regression fit line should not be visible

```python
dataset = pd.read_csv("social.csv")
sns.set(style="darkgrid")
sns.regplot(x=dataset['Age'],y=dataset['EstimatedSalary'],fit_reg=False)
```

- C) Use * symbol to show data points without the regression fit line

```python
dataset = pd.read_csv("social.csv")
sns.set(style="darkgrid")
sns.regplot(x=dataset['Age'],y=dataset['EstimatedSalary'],marker="*",fit_reg=False)
```

### 5. Scatter plot age vs estimated salary vs purchased on dataset ‘social network.csv’

```python
dataset = pd.read_csv("social.csv")
sns.lmplot (x='Age',y='EstimatedSalary',data=dataset,fit_reg=False, hue='Purchased',legend=True, palette="Set1")
```
